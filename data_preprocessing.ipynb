```python
# 1. Loading and Cleaning
df = pd.read_csv(r"C:\Users\Sneha Malale\OneDrive\Desktop\AQIProject\data\air_quality_historical.csv")
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
```

```python
# Drop initial rows with no AQI data
df = df.dropna(subset=['us_aqi']).reset_index(drop=True)
```

```python
df
```

```python
df['target_aqi_tomorrow'] = df['us_aqi'].shift(-1)
```

```python
# Lags for the current day and previous days
df['aqi_lag_1'] = df['us_aqi'] # Today
df['aqi_lag_2'] = df['us_aqi'].shift(1) # Yesterday
df['aqi_lag_3'] = df['us_aqi'].shift(2) # Day before yesterday
```

```python
# Time features
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.dayofweek
```

```python
# Drop the last row because it won't have a 'target_aqi_tomorrow'
# and drop the first few rows because they won't have all lags.
df_model = df.dropna().copy()
```

```python
print(f"Dataset size after cleaning and lag creation: {len(df_model)}")
print(df_model[['date', 'us_aqi', 'aqi_lag_1', 'aqi_lag_2', 'aqi_lag_3', 'target_aqi_tomorrow']].head())
```

```python
# Save the preprocessed data for reference
df_model.to_csv('mumbai_aqi_preprocessed.csv', index=False)
```
